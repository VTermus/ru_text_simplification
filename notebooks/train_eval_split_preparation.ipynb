# Preparation of train and evaluation split datasets

```python
import pandas as pd
import re

df = pd.read_csv("data/source_target_sentences.csv", sep=";", encoding='cp1251')
```

```python
# filter by target length
df_filtered = df[
    df["target"].apply(lambda x: len(re.findall(r"\b\w+\b", str(x))) > 6)
]

print(df_filtered)
```

```python
# create training and evaluation subsets
train_df = df_filtered.iloc[:10000]
eval_df = df_filtered.iloc[10000:]

train_df.to_csv("train_source_target_sentences.csv", index=False, sep=",")
eval_df.to_csv("eval_source_target_sentences.csv", index=False, sep=",")
```

```python
import pandas as pd
import re

df = pd.read_csv("data/eval_source_target_sentences.csv", sep=",", encoding='utf8')

print(len(df))
```

```python
i = len(df) // 3

print(i)
```

```python
# divide the evaluation set into three equal subsets
a_df = df.iloc[:i]
b_df = df.iloc[i:i*2]
c_df = df.iloc[i*2:]


a_df.to_csv("a_eval_source_target_sentences.csv", index=False, sep=",")
b_df.to_csv("b_eval_source_target_sentences.csv", index=False, sep=",")
c_df.to_csv("c_eval_source_target_sentences.csv", index=False, sep=",")
```
