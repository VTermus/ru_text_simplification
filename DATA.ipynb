```python
import pandas as pd
import re

df = pd.read_csv("/content/source_target_sentences.csv", sep=";", encoding='cp1251')

import re

df_filtered = df[
    df["target"].apply(lambda x: len(re.findall(r"\b\w+\b", str(x))) > 6)
]

print(df_filtered)

train_df = df_filtered.iloc[:10000]
eval_df = df_filtered.iloc[10000:]

train_df.to_csv("train_source_target_sentences.csv", index=False, sep="\t")
eval_df.to_csv("eval_source_target_sentences.csv", index=False, sep="\t")
```
